## Transformation

- sales: sls_ord_num, order_date, ship_date, due_date, sales, quantity, price
- prduct: product_number
- customer: customer_id

```python
query = """
SELECT
    crm.sls_ord_num AS order_number,
    pr.product_number,
    cu.customer_id,
    crm.order_date,
    crm.ship_date,
    crm.due_date,
    crm.sales AS sales_amount,
    crm.quantity,
    crm.price
FROM silver.crm_sales crm
LEFT JOIN gold.dim_products pr
    ON crm.product_number = pr.product_number
LEFT JOIN gold.dim_customers cu
    ON crm.customer_id = cu.customer_id;
"""
df = spark.sql(query)
```

```python
df.limit(10).display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
```
