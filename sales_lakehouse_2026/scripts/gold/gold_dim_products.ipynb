## dim_products
- crm_product: product_id, product_number, product_name, product_cost, product_line, start_date, end_date, category_id
- px_cat: category_id, category, subcategory, maintenance_flag
dim_products: product_id, roduct_number, product_name, product_cost, product_line, start_date, end_date, category_id, category, subcategory, maintenance_flag

```python
query = """
    SELECT 
    crm.product_id,
    crm.product_number,
    crm.product_name,
    crm.category_id,
    erp.category,
    erp.subcategory,
    crm.product_line,
    crm.start_date,
    erp.maintenance_flag
    FROM silver.crm_product crm
    JOIN silver.erp_px_cat_g1v2 erp
    ON crm.category_id = erp.category_id
    LIMIT 50
"""

df = spark.sql(query)
```

```python
df.limit(10).display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```
